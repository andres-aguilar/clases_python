## Ejemplos de su WEB

```python
"""
Simple demo of a horizontal bar chart.
"""
import matplotlib.pyplot as plt; plt.rcdefaults()
import numpy as np
import matplotlib.pyplot as plt


# Example data
people = ('Tom', 'Dick', 'Harry', 'Slim', 'Jim')
y_pos = np.arange(len(people))
performance = 3 + 10 * np.random.rand(len(people))
error = np.random.rand(len(people))

plt.barh(y_pos, performance, xerr=error, align='center', alpha=0.4)
plt.yticks(y_pos, people)
plt.xlabel('Performance')
plt.title('How fast do you want to go today?')

plt.show()
```

```python
"""
Demo of the errorbar function.
"""
import numpy as np
import matplotlib.pyplot as plt

# example data
x = np.arange(0.1, 4, 0.5)
y = np.exp(-x)

plt.errorbar(x, y, xerr=0.2, yerr=0.4)
plt.show()
```

```python
"""
Demo of scatter plot on a polar axis.

Size increases radially in this example and color increases with angle (just to
verify the symbols are being scattered correctly).
"""
import numpy as np
import matplotlib.pyplot as plt


N = 150
r = 2 * np.random.rand(N)
theta = 2 * np.pi * np.random.rand(N)
area = 200 * r**2 * np.random.rand(N)
colors = theta

ax = plt.subplot(111, polar=True)
c = plt.scatter(theta, r, c=colors, s=area, cmap=plt.cm.hsv)
c.set_alpha(0.75)

plt.show()
```

```python
"""
Demo of scatter plot on a polar axis.

Size increases radially in this example and color increases with angle (just to
verify the symbols are being scattered correctly).
"""
import numpy as np
import matplotlib.pyplot as plt

plt.style.use('ggplot')  # usar el estilo de ggplot
N = 150
r = 2 * np.random.rand(N)
theta = 2 * np.pi * np.random.rand(N)
area = 200 * r**2 * np.random.rand(N)
colors = theta

ax = plt.subplot(111, polar=True)
c = plt.scatter(theta, r, c=colors, s=area, cmap=plt.cm.hsv)
c.set_alpha(0.75)

plt.show()
```

##Combinar pandas con matplotlib para generar gr√°ficas con nuestros datos

```python
import matplotlib.pyplot as plt
plt.style.use('ggplot')  # usar el estilo de ggplot
from numpy.random import randn
import pandas as pd
```

```python
df = pd.DataFrame({'a': randn(1000) + 1, 'b': randn(1000),
                 'c': randn(1000) - 1}, columns=['a', 'b', 'c'])
```

```python
df.head()
```

```python
# le indicamos al DataFrame que vamos a graficar un histograma con una transparencia de 0.5
fig, ax = plt.subplots(figsize=(8,5))
ax = df.plot(kind='hist', alpha=0.5, title="Titulo", ax=ax)

ax.set_ylabel('Y label')  # Etiqueta del eje y
ax.set_xlabel('X label')  # Etiqueta del eje x

fig.tight_layout()  # empaquetar la gr√°fica
```

```python
# plt.savefig("fig1.svg", format='svg')  # podemos guardar la gr√°fica 
plt.show() # o simplemente mostrarla
```

```python
plt.close()
```

```python
fig, ax = plt.subplots(figsize=(8,5))  # creamos una nueva figura (fig = figura, ax = lienzo)

a = ax.hist(df["a"], bins=10, histtype="stepfilled", alpha=0.5, label="a") 
b = ax.hist(df["b"], bins=10, histtype="stepfilled", alpha=0.5, label="b")  
c = ax.hist(df["c"], bins=10, histtype="stepfilled", alpha=0.5, label="c")  

ax.legend()

plt.title('TITULO')
plt.xlabel('ETIQUETA DE X')
plt.ylabel('ETIQUETA DE Y')

plt.savefig("fig1.png", format='png')
plt.close()
```
